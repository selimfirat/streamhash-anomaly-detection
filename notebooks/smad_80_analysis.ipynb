```python
import pandas as pd
import json
import pickle

df = pd.DataFrame(columns=["dataset", "algorithm" , "ap"]) # "params", runned_in_secs
algorithms = ["ocsvm", "shocsvm", "iforest", "shiforest"]
for alg in algorithms:
    scores = pickle.load(open("../pkl/1537566095.2590454_"+ alg + ".pkl", "rb"))
    for key, score in scores.items():
        for s in score["scores"]:
            s["params"] = score["params"]
    for key, score in scores.items():
        for s in score["scores"]:
            df = df.append({
                "algorithm": alg,
                "dataset": s["dataset"],
                "ap": s["ap"],
            }, ignore_index=True)
```

```python
xstream_results = [{
    "algorithm": "xstream",
    "dataset": "cancer",
    "ap": 0.845,
},
{
    "algorithm": "xstream",
    "dataset": "ionosphere",
    "ap": 0.848,
},{
    "algorithm": "xstream",
    "dataset": "telescope",
    "ap": 0.344,
},{
    "algorithm": "xstream",
    "dataset": "indians",
    "ap": 0.216,
},{
    "algorithm": "xstream",
    "dataset": "gisette",
    "ap": 0.090,
},{
    "algorithm": "xstream",
    "dataset": "isolet",
    "ap": 0.112,
},{
    "algorithm": "xstream",
    "dataset": "letter",
    "ap": 0.122,
},{
    "algorithm": "xstream",
    "dataset": "madelon",
    "ap": 0.097,
},
]

for res in xstream_results:
    df = df.append(res, ignore_index=True)
```

```python
idx = df.groupby(["algorithm", "dataset"])['ap'].transform(max) == df['ap']
df = df[idx]
df = df.sort_values(ascending=False, by=["dataset", "ap"])
from IPython.display import display, HTML
display(HTML(df.to_html(index=False)))
```

```python
res_df = pd.DataFrame(columns=["dataset", "ocsvm", "iforest", "shocsvm", "shiforest", "xstream"])
res_dict = {}
for k, r in df.iterrows():
    if r[0] in res_dict.keys():
        res_dict[r[0]][r[1]] = r[2]
    else:
        res_dict[r[0]] = {
            "dataset": r[0],
            r[1]: r[2]
        }
```

```python
for k, v in res_dict.items():
    res_df = res_df.append(v, ignore_index=True)
```
